### Batch Learning: 
System is incapable of learning from incremental data. 

#### Offline Learning - System is trained with the data and put into production and system applies what it has learnt without learning further. 

Downside of offline learning is every time new data is fed in the system has to learn from scratch along with the old data. This is not suitable for rapidly changing environments as stock prices. In this case we need a reactive solution

Offline learning can be computationally expensive and not suitable when the amount of data coming in is voluminous 

#### Online learning is a better option for learning incrementally 
	Data fed in sequences or chunked by small groups called mini-batches

Better for continuously flowing data and adapt to change rapidly or autonomously 

Online learning algorithms can be used for huge data sets which cannot be fit into a computers main memory- out-of-core learning. 

	- Algorithm loads part of the data
	- Runs the training step on the data 
	- Repeats the process until it has run on all of the data

Warning- out-of-core learning is usually done offline 

Parameter of online learning is - learning rate - how fast the algorithm should adapt to changing data. 

High learning rate will make the system to adapt more quickly to new data but tends to forget the old data. 
Low learning rate will make system less sensitive to the noise in the new data. 

Challenges in online learning: 
	Performance gradually declines when the bad data is fed into the system 
	
Mitigation: System should be monitored for bad data and should be rolled back to the previous state   when there is a dip in the performance.

#### Challenges in Machine Learning 

Bad data 
Bad Algorithm

Unreasonable effectiveness of the data - In order for an algorithm to perform well a lot of data is required as a training set. 

Nonrepresentiveness of the training data

It is crucial to use a training set that is representive of the cases you want to generalize to. Nonrepresentiveness of training data, the algorithm will provide insufficient accuracy.

##### Sampling noise - 
when the training data sampled is less. 
##### Sampling bias - 
when the training data sampled is large and do not represent very well. (if the sampling method is flawed).

### Overfitting the training data 

The algorithm / model perform well on the training data and doesn't generalize well. 

possible solutions to reduce the overfitting 
    Simply the model by selecting fewer parameters.(linear model is better than a nonlinear one) 
    Gathering more training data.
    Reduce the noise in the training data. 
 
##### Regularization 
Constraining the model to make it simpler and reduce the risk of overfitting 

The amount of regularization to apply during learning can be controlled by hyperparameter. 
##### hyperparameter 
is a parameter of a learning algorithm (not of the model). This should be set before training the model and should be constant.
    If the hyperparameter is set very large, then the slope will be close to zero and model will not overfit the data but less likely to give the best solution. 

### Underfitting the training data

The algorithm / model is too simple to learn the underlying structure of the data. 

possible solution to reduce underfitting 
    1. Selecting a powerful model, with more parameters 
    2. Feeding better features to the learning algorithm (Feature Engineering) 
    3. Reducing the constraints on the model (eg: reducing the regularization parameter) 


#### Summary

###### Machine Learning is about making machines get better at some task by learning from data, instead of having to explicitly code rules.

###### There are many different types of ML systems: supervised or not, batch or online, instance-based or model-based, and so on.

######  a ML project you gather data in a training set, and you feed the training set to a learning algorithm. If the algorithm is model-based it tunes some parameters to fit the model to the training set (i.e., to make good predictions on the training set itself), and then hopefully it will be able to make good predictions on new cases as well. If the algorithm is instance-based, it just learns the examples by heart and generalizes to new instances by comparing them to the learned instances using a similarity measure.

######  system will not perform well if your training set is too small, or if the data is not representive, noisy, or polluted with irrelevant features (garbage in, garbage out). Lastly, your model needs to be neither too simple (in which case it will underfit) nor too complex (in which case it will overfit).

### Testing and validating

once the model is trained we need to test the model on the new data which is not fed into the model 

In general, its good to split the data into training set and test set. 

##### Generalization error
Error rate we obtained when running a model on the new cases.
Estimate the error rate.

Note if the training error is low but the generalization error is too high, the model has overfitted the data 

Its common to use 80% training data and 20 test data. this depends on the data size 

#### Hyperparameter tuning and model selection 


#### Hold-out Validation

holding out the part fo the training set and evaluate several candidate models and select the best one. 

New held out set is called the Validation set. (dev set or development set) 
Now we train muliple models with various hyperparameters on the reduced training set (full traininset - validation set) 
After the hold out validation process, you train the best model on the full training set. 
Finally evaluate the model on the test set to get the estimate of the generalization error. 

##### Validation set too small:
The above works mostly unless the validation set is too small. The model evaluation is imprecise.
sub-optimal model might be selected by mistake. 

##### Validation set too large: 
When the model will be trained on the full training set, its not ideal to compare candidate models  trained on much smaller training set. 
eg: its like selecting the fastest sprinter for a marathon. 

solution to solve the above problem: Perform repeated cross-validation using small validations sets. 

Each model is evaluated once per the validation set, after it is trained on the rest of the data. By averaging out evaluations of the model, we get a much more accurate measure of the performance. The drawback is the training time is multiplied by the number of validations sets. 

#### Data Mismatch

Ensure the data is representive of the validation set and the training set. 

if the validation set yields poor performance of the model. Create another validation set called train-dev test. Train the model on the training set, evaluate it against the train-dev set and if the model perform well run it on the validation set. If the validation set yields poor performance then the issue is with data mismatch. 

If the model performs poorly on the train-dev set then we have overfitting issue. Need to regularize the model.
